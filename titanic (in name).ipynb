```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#read_csv
train_df = pd.read_csv('train.csv')
train_df.head()
test_df = pd.read_csv('test.csv')
test_df.head()
# データを読み込む関数の定義

def load_dataset():
    train_df = pd.read_csv('train.csv')
    test_df = pd.read_csv('test.csv')
    labels = train_df.Survived
    return train_df, test_df, labels

train_df, test_df, labels = load_dataset()
```

```python
def make_dummies(df):
    sex_dum = pd.get_dummies(df['Sex'])
    df = pd.concat((df,sex_dum),axis=1)
    #元のデータを消す
    df = df.drop('Sex',axis=1)#axis=1が列
    
    #次にEmbarkedもダミー変数にする。
    emb_dum = pd.get_dummies(df['Embarked'])
    df = pd.concat((df,emb_dum),axis=1)
    #元のデータを消す
    df = df.drop('Embarked',axis=1)#axis=1が列
    return df
```

```python
#ダミー変数に変換
train_df =make_dummies(train_df)
test_df =make_dummies(test_df)
```
