```python
# pandasを利用しないで読み込み(品種カウント)
count_dict = {}
with open('./stationery.csv') as f:
    for idx, row in enumerate(f):
        if idx == 0:
            continue
        colmuns = row.rstrip().split(',')
        name = colmuns[0]
        if name in count_dict:
            count_dict[name] += 1
        else:
            count_dict[name] = 1

for key, value in count_dict.items():
    print(key, value)
```
